```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_excel('list_of_thesis.xlsm')
train.head()
```

```python
Data = train[['키워드(한국어)']]
data = Data.drop_duplicates().dropna()

len(data)
```

```python
data_ = data.values.tolist()

data_[:5]
```

```python
import re
regex = []

for i in range(len(data)):
   text = re.sub('[^0-9r-힣]', ' ', str(data_[i]))
   text = re.sub('[a-zA-Z]' , ' ', text)
   text = re.sub(' +', ' ',text)
   regex.append(text)
```

```python
regex[:5]
```

```python
!pip install konlpy
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
noun = []

for j in range(len(regex)):
  noun.append(okt.nouns(regex[j]))
  print(str(j)+ "번째 실행중")
```

```python
noun[:384]
```

```python
import gensim

import warnings
warnings.filterwarnings('ignore')
```

```python
id2word = gensim.corpora.Dictionary(noun)
```

```python
texts = noun

corpus = [id2word.doc2bow(text) for text in texts]

print(corpus[:1])
```

```python
print(corpus)
```

```python
number = 5
```

```python
model = gensim.models.ldamodel.LdaModel(
    corpus = corpus,
    id2word = id2word,
    num_topics = number,
    random_state=1025
)
```

```python
model.show_topics(formatted=False)
```

```python
!pip install pyLDAvis==2.1.2
```

```python
import pyLDAvis.gensim

vis = pyLDAvis.gensim.prepare(model, corpus, id2word)
```

```python
pyLDAvis.save_html(vis, '키워드 분석(한국어).html')
```
