```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('list_of_thesis.xlsm')

print(df.shape)
```

```python
df.head()
```

```python
train = df['논문명']
```

```python
train.head()
```

```python
Ntrain = train.drop_duplicates().dropna()
```

```python
len(Ntrain)
```

```python
Ntrain
```

```python
Ntrain_ = Ntrain.tolist()
```

```python
Ntrain_
```

```python
import re
regex = []

for i in range(len(Ntrain_)):
   text = re.sub('[^0-9r-힣]', ' ', str(Ntrain_[i]))
   text = re.sub(' +', ' ',text)
   regex.append(text)
```

```python
Ntrain_[:10]
```

```python
!pip install konlpy
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
noun = []

for j in range(len(regex)):
  noun.append(okt.nouns(regex[j]))
  print(str(j)+ "번째 실행중")
```

```python
noun[:2]
```

```python
import gensim

import warnings
warnings.filterwarnings('ignore')
```

```python
id2word = gensim.corpora.Dictionary(noun)
```

```python
texts = noun

corpus = [id2word.doc2bow(text) for text in texts]

print(corpus[:1])
```

```python
print(corpus)
```

```python
number = 15
```

```python
model = gensim.models.ldamodel.LdaModel(
    corpus = corpus,
    id2word = id2word,
    num_topics = number,
    random_state=1025
)
```

```python
model.show_topics(formatted=False)
```

```python
!pip install pyLDAvis==2.1.2
```

```python
import pyLDAvis.gensim

vis = pyLDAvis.gensim.prepare(model, corpus, id2word)
```

```python
pyLDAvis.save_html(vis, 'result.html')
```
